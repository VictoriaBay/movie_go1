## 5.3 Exploring data

```python
#Accessing movie_basics table
movie_basics_df = pd.read_sql(""" SELECT * FROM movie_basics;""",conn)
movie_basics_df.head(5)
```

```python
#Accessing movie_ratings table
movie_ratings_df = pd.read_sql(""" SELECT * FROM movie_basics;""",conn)
movie_ratings_df.head(5)
```

```python
#Joining 2 tables with columns that we will analyze
basics_with_ratings = pd.read_sql("""
SELECT movie_basics.movie_id , movie_basics.genres, movie_ratings.averagerating, movie_basics.runtime_minutes, movie_basics.original_title
FROM movie_basics 
JOIN movie_ratings ON movie_ratings.movie_id = movie_basics.movie_id ORDER BY movie_ratings.averagerating DESC;""",conn)
basics_with_ratings_df = pd.DataFrame(basics_with_ratings)
basics_with_ratings_df = basics_with_ratings_df.rename(columns ={'original_title':'movie'})
basics_with_ratings_df
```

```python
#Adding frofit column
budgetTn['profit'] = budgetTn['worldwide_gross']- budgetTn['production_budget']
budgetTn
```

```python
movie_budget = budgetTn.merge(tmdbMovies, on=["movie"]) 
movie_budget.head(5)
```

```python
main_df = movie_budget.merge(basics_with_ratings_df, on=['movie'])
mainDF = main_df[["movie", "genres", "averagerating", "runtime_minutes", "production_budget", "domestic_gross", "worldwide_gross", "profit"]]
mainDF
```

```python
mainDF = mainDF.sort_values('profit', ascending= False)
mainDF.head(10)
```

```python
top10_dom = mainDF.nlargest(10,'domestic_gross')[['domestic_gross','movie', 'runtime_minutes','genres','worldwide_gross', 'averagerating']]
top10_dom.set_index('movie')
```

```python
sns.barplot(data = top10_dom, x = 'domestic_gross', y = 'movie', hue = 'runtime_minutes', dodge = False)
plt.title('Movies with highest domestic Gross', fontsize = 16)
plt.xlabel('Domestic Gross', fontsize = 15)
plt.ylabel('Movie Titles', fontsize = 15)
plt.legend(bbox_to_anchor = (1.05,1), loc = 2)
plt.show();
```

```python
sns.barplot(data = top10_dom, x = 'worldwide_gross', y = 'movie', hue = 'averagerating', dodge = False)
plt.title('Movies with highest worldvide gross', fontsize = 16)
plt.xlabel('profit', fontsize = 15)
plt.ylabel('Movie Titles', fontsize = 15)
plt.legend(bbox_to_anchor = (1.05,1), loc = 2)
plt.show();
```
